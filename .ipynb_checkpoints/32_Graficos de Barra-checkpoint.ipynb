```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randn(1000)
```

```python
x
```

```python
plt.hist(x)
plt.show()
```

```python
plt.hist(x, bins=25) #Bins -> Subdividiu os Intervalos em 25 Barrinhas.
plt.show()
```

```python
plt.hist(x,bins=10 ,color='g')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title("Grafico de Barras", size=20)
plt.show()
```
