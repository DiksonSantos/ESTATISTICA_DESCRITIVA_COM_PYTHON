# GRÁFICOS DE BARRAS OU COLUNAS:

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
y = [ 2,5,2,7,5,1]
x = ['N1','N2','N3','N4','N5','N6']
X2 = ['VARIAVEL_Um','VARIAVEL_Dois','VARIAVEL_Três','VARIAVEL_Quatro','VARIAVEL_Cinco','VARIAVEL_Seis']
```

```python
plt.barh(X2,y, color='g')
plt.title('All Green')
plt.show()
```

```python
plt.bar(x,y, color='black')
plt.xlabel('Eixo -> X')
plt.ylabel('Eixo -> Y')
plt.show()
```
