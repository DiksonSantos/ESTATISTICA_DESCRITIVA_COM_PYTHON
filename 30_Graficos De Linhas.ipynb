```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
y = [ 6,8,3,1,9]

x = [1,2,3,4,5]

X2 = ['Seg', 'Ter', 'Qua', 'Qui', 'Sex']
```

```python
plt.plot(x, y, 'o-g') #y de Yellow , g = Green

plt.xlabel('Eixo X', size=20)
plt.ylabel('Eixo Y', size=20)
plt.title('Grafico Aqui', size=20)
plt.show()
```

```python
plt.plot(X2, y, 'o-b') #y de Yellow , g = Green

plt.xlabel('Eixo X', size=20)
plt.ylabel('Eixo Y', size=20)
plt.title('Grafico Aqui', size=20)
plt.show()
```
